```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt


file = r'...\arquivo.xls' # endere√ßo de onde esta o arquivo xlsx ou csv
txt = pd.read_excel(file)
clould = str(txt)

wordcloud = WordCloud(background_color="black", width=1600, height=800).generate(clould)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis("off")
wordcloud.to_file(r" ...\nome_imagem.png") # endere√ßo para salvar a imagem
plt.show()
```
